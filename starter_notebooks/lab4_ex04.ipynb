# 1. Przykład 4

W tym przykładzie wykorzystamy API wysokiego poziomu o nazwie Keras zintegrowane z pakietem TensorFlow (https://www.tensorflow.org). Zastosujemy metody optymalizacji na podstawie spadku wzdłuż gradientu.

Na początku wskażmy pakiety, które będziemy wykorzystywać w projekcie

```python
import numpy as np
import tensorflow as tf
```

oraz wygenerujmy przykładowe, losowe dane.

```python
# Parametr określający ilość danych zmiennej niezależnej x
dataset_training_size = 1000
# Prawdziwe wartości parametrów
Wt = 0.2
bt = 0.4


# Dodanie szumu do danych
# noise = 0
noise = np.random.rand(dataset_training_size).astype(np.float32)/20

# Założony model:  y = Wt* x  + bt + noise
# Generacja danych zmiennej niezależnej x
x_t = np.linspace(0, 1, dataset_training_size).astype(np.float32)

# Tu można wygenerować wartości nierównomiernie
# x_t = np.random.rand(dataset_training_size).astype(np.float32)

x_training = x_t.reshape(dataset_training_size, 1)

# Sortowanie danych, jak nie zostały stworzone liniowo
# x_training = (np.sort(x_t)).reshape(dataset_training_size,1)

W_training = np.array([Wt])

# Wyliczenie prawdziwych wartości modelu y
y_t = (np.dot(x_training, W_training.T)+bt+noise).astype(np.float32)
y_training = np.transpose(np.array([y_t]))
```

Definicja modelu będzie dziedziczyła po klasie Model, a następnie następuje zdefiniowanie parametrów i funkcji modelu prostej regresji liniowej.

```python
# Użycie Keras High Level API do rozwiązania problemu

# 1. Definicja modelu
class SimpleKerasModel(tf.keras.Model):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    self.n = 1
    self.W = tf.Variable(tf.ones([self.n, 1]), name="W")
    self.b = tf.Variable(tf.ones([1]), name="b")

  def __call__(self, x, **kwargs):
    return self.W * x + self.b
```

Ostatecznie należy skonfigurować proces uczenia, ustawić algorytm optymalizacyjny, funkcję kosztu oraz wywołanie procedury treningu dla określonych parametrów procedury

```python
# Wywołanie modelu
keras_model = SimpleKerasModel()

# Ustawienie hiperparametrów
learning_rate = .007
epochs = 100
batch_size=50

# Ustawienie parametrów treningu
keras_model.compile(
    # Użycie trybu Eager
    run_eagerly=True,
    # wybór optymalizatora
    optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate),
    # wybór funkcji straty
    loss=tf.keras.losses.mean_squared_error
)

# 5. Trening modelu z ustawionymi hiperparametrami
keras_model.fit(x_training, y_training, epochs=epochs, batch_size=batch_size)

print("\nEstimated W: ", keras_model.W.numpy())
print("\nEstimated b: ", keras_model.b.numpy())
```
